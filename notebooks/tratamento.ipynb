<h2>Tratamentos necessários identificados:
    - Remover colunas sem nome e dados;
    - Trocar o tipo de dado da coluna DATA;
    - Remover colunas nulas (PR_ID == 107).

Importando pandas e lendo o arquivo csv

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw.csv", delimiter=";")
```

Removendo colunas vazias

```python
df = df.drop(columns=["Unnamed: 10","Unnamed: 11","Unnamed: 12","Unnamed: 13"])
```

```python
df.head()
```

Trocar o tipo de dado da coluna DATA

```python
df["DATA"] = pd.to_datetime(df["DATA"], format="%d/%m/%Y")
df.info()
```

Trocando o tipo de dado das colunas de id para string - Optei em fazer para evitar calculos com colunas de id, que devem ser únicas.

```python
x = ["CO_ID","CL_ID","PR_ID"]

for i in x:
    df[i] = df[i].astype("str")
```

Removendo colunas nulas

```python
df[df["PR_ID"]=="107"]
```

```python
df = df.drop(df.loc[df["PR_ID"]=="107"].index)
df.shape
```

Exportando o dataset processado

```python
# Organizando o índice depois de todo tratamento
df = df.reset_index(drop=True)
```

```python
#Exportação
df.to_csv("../data/processed.csv", index=False)
```
