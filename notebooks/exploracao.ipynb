```python
# Verificando duplicatas
df[df.duplicated()].shape
```

Percebe-se um alto valor de linhas duplicadas. Porém, interpretando o contexto do dataset (varejo) e a ausência de uma coluna de quantidade, as linhas duplicadas parecem representar essa quantidade do produto comprado que falta no dataset. Por exemplo, duas linhas exatamente iguais com o produto (PR_NOME) banana, parece representar que o cliente comprou duas unidades de banana (cacho). Logo, não removerei essas duplicatas.

```python
# Verificando valores distintos de colunas categóricas
print(df["CL_GENERO"].value_counts())
print(df["CL_EC"].value_counts())
print(df["CL_SEG"].value_counts())
print(df["PR_CAT"].value_counts())
```

Nota-se valores #N/D nas colunas PR_CAT e PR_NOME, mas não estão sendo considerados nulos por que está escrito #N/D como um valor.

```python
# df[df["PR_CAT"]=="#N/D"]
# df[df["PR_NOME"]=="#N/D"]
# df[df["PR_ID"]==107]
print(f"Coluna PR_CAT com valores #N/D: {df[df["PR_CAT"]=="#N/D"].shape}")
print(f"Coluna PR_NOME com valores #N/D:{df[df["PR_NOME"]=="#N/D"].shape}")
print(f"Coluna PR_ID com id 107 : {df[df["PR_ID"]==107].shape}")
```

Perceb-se que os valores "nulos" estão nas linhas onde PR_ID é igual a 107

```python
# Calculando a porcentagem que os valores nulos identificados representam para o dataset
x = df[df["PR_ID"]==107].shape[0] / df.shape[0]
print(f"Porcentagem de nulos: {x *100:.2f}%")
```

Como os valores nulos representam apenas 0,44% de todo o dataset, serão removidos para que não atrapalhem as futuras análises
